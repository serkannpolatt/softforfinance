# Generative AI

2022'nin başına kadar, sınıflandırıcı ve tahmin/gerileme sistemleri gibi ayrımcı yapay zekanın nasıl çalıştığını görmüştük. Bir dizi veri noktası eğitimi verildiğinde, model öngörülemeyen veri noktalarını birbirinden ayırabilir. Birdenbire üretken yapay zeka etrafındaki heyecan arttı ve önceden var olan geniş verilerden yeni içerik üretme kavramı galip geldi. Gerçek ve sentetik veriler üretmek için makine öğrenimi modellerinden para kazanmayı muhtemelen kim hayal edebilirdi? Derin Öğrenme ve NLP, RNN'lerden, LSTM'lerden, kodlayıcı-kod çözücüden ve VAE modellerinden GAN'lara (Generative Adversarial Networks) kadar uzun bir yol kat etti. AGI (Yapay Genel Zeka) alanı moda, otomotiv, finans, sağlık, eğitim ve daha pek çok sektörde devrim yaratıyor. Yıllar geçtikçe araştırmacılar, insan beynini taklit edecek sinir ağları fikrini başlattılar; bu kadar ilerlemenin ardından bugün, bu YGZ modellerinde insan benzeri konuşmalar görebiliyoruz. Yapay zekanın tıpkı Shakespeare gibi şiir ve edebiyat yazabilmesi ya da Mars'ta ata binen bir astronotun resmini ortaya çıkarabilmesi büyüleyici. 

Generative, teknoloji dışı insanlara büyük bir destek sağladı. Artık yapay zekayı uygulamak için programlamayı öğrenmeye gerek yok. Kodlayıcılar/programcılar için de kod oluşturma modelleri bir nimettir. Yalnızca kod üretimi değil, aynı zamanda Python'dan Java'ya, Java'dan C++'ya ve çok daha fazlasına kod özetleme gibi bir programlama dilinden diğerine kod çevirisi. Her ne kadar araştırmacılar ve uygulayıcılar için bu tam tersi ve en son gelişmelerin hızına ayak uydurmak oldukça zorlayıcı olsa da. Çağrı merkezleri, müşteri destek yöneticileri ve içerik oluşturma/yaratma şirketleri gibi birçok iş, otomasyon aşamasını gözlemliyor. Yapay zeka tarafından üretilen sesli asistanlar, 3 boyutlu görüntüleme, AR/VR, meta veri dizisi, robot teknolojisi ve sürücüsüz araçlar, dünyayı daha yeni olanaklarla değiştirebilir. Sanat ve tasarım, tarihten sahneleri yeniden canlandırabilen, bir resimde insanları zehirleyen, resimlerdeki insanları tepeden denize kadar yeni arka planlara götürebilen veya kıyafetlerini değiştirebilen difüzör modelleri ile hayal edilemeyecek şeylerin gerçekleştiğini gördü. Model eğitimi için sentetik verilerin oluşturulması, gerçek zamanlı benzer verilerin üretildiği ve dolayısıyla gerçek dünya verilerinin gizliliğini koruyan bir başka önemli husus haline geldi. Filmlerin manzarası, fanteziler ve diğer görsel düzenleme seçenekleriyle hızla daha büyüleyici hale gelebilir. Oturup uzun makaleler veya ders kitapları okumak zorunda kaldığımız günler geride kaldı; bunu modele besleyebilir ve onun bir özet söylemesini veya belirli bölümlerden sorular sormasını sağlayabiliriz.

# Foundation Models


Credits: AWS

Burada geleneksel makine öğrenimi modellerinin nasıl davrandığını görüyoruz. Denetimli öğrenmeyle elde edilen büyük miktardaki etiketli veriler üzerinde eğitilirler ve daha sonra görüntü tanıma, duygu analizi vb. gibi belirli görevleri gerçekleştirmek için konuşlandırılırlar. Bu modeller yalnızca üzerinde eğitim aldıkları görevler hakkında tahminlerde bulunabilir ve daha yeni olasılıklar ortaya koyamazlar. Yalnızca Regresyon ve Sınıflandırma görevleriyle sınırlıdır. Temel modeller son birkaç yılda gelişmeye başladı ve çok büyük miktarda etiketlenmemiş veri üzerinde eğitildiğinden farklıdır, dolayısıyla kendi kendini denetleme veya denetimsiz öğrenmeden öğrenir ve bu da daha geniş modellere giden yolları açar. Sistem, ister ayrımcı ister üretken olsun, her türlü göreve uyum sağlayabilir. 

Temel Modeli terimi, İnsan Merkezli Yapay Zeka Enstitüsü'nün (HAI) Temel Modeller Araştırma Merkezi (CRFM) tarafından icat edildi ve bu terim, büyük ölçekli veriler üzerinde kendi kendini denetleyen veya yarı denetimli yöntemlerle eğitilen herhangi bir modelin uyum sağlayabileceğini söyledi. kendisini bilişsel aşağı akışlı alana özgü görevlere yönlendirir. Temel modellerinin iki temel özelliği vardır; ortaya çıkma ve homojenleşme. Burada verinin etiketlenmesinde insan yer almıyor, modelin açıkça kodlanması yerine ortaya çıkması gerekiyor. Model, aksi halde beklenmeyen özellikleri sergileyebilir. Örneğin geniş dil veri seti üzerinde eğitilen böyle bir model, açıkça programlanmasına gerek kalmadan kendi hikayelerini üretebilir, aritmetik yapabilir. Ek olarak, modelin bu özelliklerinin, alt ölçeklendirme yasalarındaki ihlaller nedeniyle önceden çıkarımlarda bulunulması bazen zor olabilir. Homojenizasyona gelmek, aynı yöntemin gelişmiş görevler için birden fazla alanda uygulanması anlamına gelir ancak aynı zamanda "tek başarısızlık noktası" niteliğine de sahip olabilir.

OpenAI GPT4, Dall-E2 ve StabilityAI Stable Diffusion en çok ilgi gören FM'ler oldu. ChatGPT gibi bir temel model oluşturduğunuzu ve bunun içindeki adımların neler olduğunu hayal edin. Diğer tüm makine öğrenimi problemlerinde olduğu gibi bu da veri toplama, hazırlama vb. ile başlayacaktır. En baştan bir FM oluşturmak için en önemli faktör gerekli donanım hesaplaması olacaktır. Mevcut tüm FM'lerin, büyük miktardaki veri ve parametreler için dağıtılmış GPU'lar üzerine inşa etmek için birkaç ayı vardır. İlgili adımların neler olduğunu görelim.

* Temel dil modelini oluşturmak için ön eğitim. Maksimum sürenin söz konusu olduğu yer burasıdır. Ön eğitim, süper bilgisayarlarla, çoğunlukla Microsoft Azure, AWS veya GCP gibi bir bulut hizmet sağlayıcısıyla büyük ölçekli verileri içerir. Bu veriler internetin herhangi bir yerinden, örneğin Wikipedia'dan, milyonlarca veya milyarlarca veri noktası içeren açık kaynaklı veri kümesi platformlarından, GitHub kodları, kitaplar ve arXiv, C4 ve CommonCrawl'dan geliyor olabilir.

* Bu da bizi her geçen gün daha fazla parametre sayısına sahip yeni bir modelin ortaya çıktığını gördüğümüz başka bir düşünceye getiriyor. Modeller parametre sayısıyla değerlendirilemez. Eğitim ne kadar fazla olursa, parametre sayısına bakılmaksızın model o kadar iyi olur.

* Temel modeller, hızlı mühendislik, denetimli ince ayar (SFT), ödül modelleme (RM) ve insan geri bildirimiyle takviyeli öğrenme (RLHF) ile daha iyi hale getirilebilir. Hızlı mühendisliğin, dil modellerinden daha iyi sonuçlar elde etmenin en basit ve etkili yolu olduğu kanıtlanabilir. Çeşitli yönlendirme tekniklerini tartışan özel bir bölümümüz var. Transformer tabanlı modeller tokenizasyon üzerinde çalışır. 

* Bir sonraki bölümde Yüksek Lisans'lara bakalım ve çeşitli teknikler kullanarak Yüksek Lisans'ın performansını artırmanın çeşitli yollarını tartışalım.

# Large Language Models

Büyük Dil Modelleri Temel Modellerin bir parçasıdır. Bunlar esas olarak, genellikle milyarlarca parametreden oluşan devasa miktardaki metin külliyatı üzerinde eğitilir. Bu sistemler insanlar gibi etkileşime girme gücüne sahiptir. Her şey Google'ın ünlü makalesiyle başladı: İhtiyacınız Olan Tek Şey Dikkat! ve modellerin doğal dili ve kelime anlamlarını, korelasyonları daha iyi anlayabildiği ve hatta çok modlu işlevler geliştirebildiği yapay zeka manzarasını tamamen değiştiren Transformer modelleri. 

Açık kaynaklı modellerin yanı sıra kapalı kaynaklı modeller de bulunmaktadır. GenAI devi OpenAI, patentli GPT modellerine sahiptir, GPT3/4 metin üretir, dilleri çevirir, farklı türde yaratıcı içerikler yazar ve sorularınızı bilgilendirici bir şekilde yanıtlar. HuggingFace BLOOM, diğer görevlerden ziyade metin oluşturmada iyi çalışır ancak uygun istem mühendisliği ile başka şeyler de yapmak için yapılabilir. En yenileri Google'ın PALM2'si, Meta AI'nın LLAMA'sı ve Cohere ile AI24Labs'ın çok daha fazlasıdır. Bu sistemler, ekipler için otomatik yanıtlarda, postalarda, kod oluşturmada (GitHub yardımcı pilotu, AWS Code Whisperer), bilgi alımında (anlamsal arama) görülen metin oluşturma (sonraki kelime/ifade tahmin edicileri) gibi metne özgü görevleri gerçekleştirmek için aşağı akışa tabi tutulabilir.


- Büyük denetlenmeyen verilerden temel modele sahip olduğumuzda, küçük bir denetlenen veri kümesi üzerinde ince ayar yapabiliriz. Parametre Verimli İnce Ayar (PEFT), bu tür sonuçların elde edilmesine yardımcı olabilir. Modeller büyüdükçe, donanım tüketimi bağımlılıkları göz önüne alındığında ince ayar yapmak zorlu bir görev haline gelir. İnce ayarlı modeller, orijinal önceden eğitilmiş modelle aynı boyutta olduğundan, depolama ve dağıtım sırasında büyük ücretler alınır. Parametre Açısından Verimli İnce Ayar (PEFT), bu darboğazların her ikisi de akılda tutularak geliştirilmiştir. PEFT, önceden eğitilmiş LLM'lerin çoğu model parametresinin dondurulmasına karşı yalnızca az sayıda (ekstra) model parametresinin ince ayarıyla ilgilenir, bu da hesaplama ve depolama maliyetlerinin azaltılmasına yardımcı olur. Bu yaklaşımın ek bir faydası da vardır; bu, LLM'lerin tam ince ayarı sırasında modellerde gözlemlenen yaygın bir davranıştır.

- Ödül modelleme, kayıp fonksiyonunun tahmin edilen ödüllerin etiketli sıralamayla tutarlılığını ölçtüğü zamandır. Ödül modeli oluşturmada nihai amaç, bir metin dizisini alabilen ve sayısal temsili içeren skaler bir ödülle geri dönen bir model veya sistem üretmektir. Bu bizi bir sonraki adıma götürür.

- Son olarak en iyi yöntem, tüm kaçınma yöntemlerini geride bırakan RLHF'dir. Bu, insan geri bildirimiyle LLM'yi optimize etmek için takviyeli öğrenmeyi kullanmanın bir yoludur. 3 ana aşamadan oluşur: Bir dil modelinin (LM) ön eğitimi, veri toplanması ve bir ödül modelinin eğitimi ve takviyeli öğrenme ile LM'ye ince ayar yapılması. RLHF, LLM'leri doğrudan insanların performansları hakkındaki değerlendirmelerinden öğrenerek eğitir. LLM'lerin aldığı tahmin veya karar yanlış veya idealin altındaysa, insanlardan gelen geri bildirimler hatayı düzeltmek veya alternatifler önermek için kullanılır. Zamanla model kendi kendine öğrenebilir ve daha iyi yanıtlar verebilir. RLHF, algoritmik çözümü tanımlamanın belirsiz olduğu ancak insanların yargısının Yüksek Lisans çıktısının kalitesini artırabildiği karmaşık görevlerde faydalıdır (örneğin, görev bir makale oluşturmaksa, insanlar Yüksek Lisans tarafından oluşturulan çıktıları puanlayabilir ve Yüksek Lisans geri bildirimi kullanır) neslini geliştirmek için).


# Prompt Engineering

İsteme dayalı ML, LLM'lere gerçekleştirmesi gereken görevi açıklayan istekler gönderir. Bunlar metinden metne veya metinden görüntüye/video/ses olabilir. Bilgi istemleri, LLM'lerle daha kolay etkileşim kurmaya yardımcı olur. İstem ne kadar kesin olursa çıktı da o kadar iyi olur. Hem istemler hem de tamamlamalar, her LLM tarafından belirlenen bağlam uzunluğunun bir parçasıdır. Modelden modele bağlı olarak ortalama 4k-8k'dir. Aşağıda farklı türdeki ipucu teknikleri verilmiştir.

* Zero-shot prompting: Bu, bir modelin açıkça eğitilmediği durumlar için yanıtlar üretebildiği zamandır. Bu, herhangi bir veri üzerinde yeniden eğitim modeli gerektirmez. Bu nedenle yeni görevin özelliklerinin farkında değildir ve genel olarak konuyla alakalı olmayabilecek yanıtlar üretir.

* Few-shot prompting: Adından da anlaşılacağı gibi modelin birkaç örnekle eğitilmesine dayanmaktadır. Bu, verilerin kolayca bulunamadığı görevler için yararlı olabilir. Bu nedenle yeni görev hakkında tamamen öğrenmenin sınırlamaları vardır.

* Chain-of-Thoughts(CoT) Prompting - Bu süreç ek bilgi sağlayarak gerekçenin anlaşılmasına yardımcı olur. Nihai cevaba nasıl ulaşılacağına ilişkin adım adım açıklama gerektiren matematik problemleri.

* Self-Consistency - Bu süreç, modelin farklı çıktılar oluşturmak için aynı komut istemi üzerinde birden çok kez çalıştırılmasına olanak tanır. Tüm cevapların çoğunluğu nihai çıktı olarak alınır.

* Tree-of-Thoughts - Bu süreç, bir sorunu çözmek için alınan kararları içeren bir ağaç oluşturur. LLM'nin karar vermedeki gerekçeyi kendi kendine değerlendirmesine ve genişlik öncelikli arama ve derinlik öncelikli aramayı kullanarak değerlendirme yapmasına ve geri izleme yoluyla keşfetmesine olanak tanır.

* LLM'lerde ince ayar yapmak, genellikle bilgi istemleri ve tamamlama biçimini kullanarak alana özgü verilerde yardımcı olur. İnce ayar, sinir ağının üst katmanlarının yeni veri kümelerinden gelen ağırlıklarla değiştirildiği eski bir yöntemdir.

# LangChain

LangChain, modellerle etkileşim kurmak, bilgi istemi şablonları ayarlamak, LLM Aracıları, zincir yanıtları ve çok daha fazlası için kullanılabilen LLM çerçevesidir. Langchain, AI24Labs, Anthropic, Cohere, HuggingFace, OpenAI ve diğer pek çok firmanın modellerinden geniş bir desteğe sahiptir. Langchain, geliştiricilerin işini kolaylaştırarak LLM uygulamaları oluşturmayı tek noktadan haline getiriyor. Şu an itibariyle Javascript ve Python desteği var. Langchain birkaç modülden oluşur: 

* Model Merkezi - Dil modelleriyle arayüz
* Veri bağlantısı - Uygulamaya özel verilerle arayüz
* Zincirler - Çağrı dizileri oluşturun
* Aracılar - Verilen üst düzey direktiflerin hangi araçları kullanacağını zincirlerin seçmesine izin verin
* Bellek - Bir zincirin çalıştırmaları arasında uygulama durumunu sürdürün
* Geri aramalar - Herhangi bir zincirin ara adımlarını günlüğe kaydedin ve yayınlayın

Langchain, geniş API kullanımı desteği ve bir Yüksek Lisans uygulaması oluşturmak için çeşitli eklentilerle popüler hale geldi. Langchain'in doldurduğu boşluğun ne olduğunu anlayalım. Diyelim ki, bankacılıkla ilgili birden fazla belgeyi ve PDF'yi okuyan ve bu belge ve PDF'lerdeki soruları yanıtlayan bir sohbet robotu oluşturmaları gerektiğini söyleyen bir bankacılık müşteriniz var. ChatGPT'ye her zaman gidemezsiniz çünkü ChatGPT'de daha önce herhangi bir bağlam belleği, bağlam uzunluğu ve belirteç uzunluğu sınırlamaları yoktur. Burası Langchain'in, zincirler veya bellek oluşturmak için vektör veritabanı indeksleri kullanılarak kaydedilen belgelerden bu bankacılık bilgilerini işleyebilen bir LLM sağlama konusunda adım attığı yerdir. Endişelenmeyin, belge yükleyiciler ve karakter parçalarına ve yerleştirmelere dönüştürme desteği de vardır. Tüm bu süreç herhangi bir boru hattına veya web uygulamasına kolayca eklenebilir ve kullanılabilir. Yaşasın!

# Use of Vector Databases, Embeddings

Vektör depoları, yüksek boyutlu vektörleri depolamak ve almak için optimize edilmiş bir veri deposu türüdür. Belge parçaları gömmelere dönüştürülür. LLM'ler çıktı olarak yüksek boyutlu vektörler üretir ve vektör depoları, bu vektörleri, dizinler oluşturularak erişilmesi verimli ve hızlı bir şekilde depolamak için kullanılabilir. Çeşitli uygulama seçenekleri mevcuttur: Milvus (Zilliz tarafından), Pinecone, Weaviate, Annoy (Spotify tarafından) ve FAISS (Meta tarafından). Bunlar, LLM belleğinin önbelleğe alınmasına veya büyük dosyalardan alana özgü verilerin hızlı bir şekilde alınmasına yardımcı olabilir. 

Vektör veritabanları, veri depolamayı ve almayı temsil etme biçimleri açısından geleneksel SQL ve NoSQL veritabanlarından farklıdır. VectorDB doğası gereği yapılandırılmış değildir; verileri ve indeksi yüksek boyutlu uzayda matematiksel vektörlerde saklama eğilimindedirler. Benzerlik aramaları gibi görevlerde daha fazla verimlilik sağlanmasına ve görüntüler, ses ve metin gibi karmaşık veri türlerinin daha iyi işlenmesine olanak tanıyan "vektörleştirme" olarak da bilinir. Gömme, bu karmaşık verileri vektörlerin veya matrislerin sayısal temsillerine dönüştürme işlemidir. Bu veri temsili biçimi, derin sinir ağı modellerinin onu anlamasını ve onunla çalışmasını kolaylaştırır. Gömmeler, boyutsallığın azaltılması, anlamsal ilişkilerin yakalanması ve benzerliğe dayalı hesaplamaların kolaylaştırılması gibi çeşitli faydalar sağlar. Öneri sistemleri, metin sınıflandırma, bilgi alma ve kümeleme gibi görevlerde yaygın olarak kullanılırlar.

Vektör veritabanlarını kullanmanın temel avantajı, vektör mesafesine dayalı olarak benzerlik arama ve almada ne kadar hızlı ve doğru olduklarıdır. Bu, tam eşleşmeleri bulmaya çalışan sorgu tabanlı veritabanlarından daha iyi çalışır; bir vektör veritabanı, anlamsal veya bağlamsal anlama dayalı olarak en benzer veya ilgili verileri alır. Vektör veritabanlarının kullanımına ilişkin kullanım stratejileri aşağıda verilmiştir:

1. Efficient Storage and Retrieval: Vektör veritabanları, yüksek boyutlu vektör verilerini verimli bir şekilde depolamak ve almak için özel olarak tasarlanmıştır. Benzerliğe dayalı sorguları hızlı ve doğru bir şekilde gerçekleştirmek için yaklaşık en yakın komşu (ANN) araması gibi özel indeksleme ve arama algoritmaları kullanırlar. Bu, onları öneri sistemleri veya görüntü/ses arama gibi benzerlik eşleşmesi gerektiren uygulamalar için çok uygun hale getirir.

2. Native Support for Vector Operations: Vektör veritabanları, mesafe hesaplamaları, vektör aritmetiği ve toplamalar dahil olmak üzere vektör işlemleri için yerel destek sağlar. Bu, karmaşık hesaplamaların doğrudan veritabanı içinde yapılmasına olanak tanıyarak, verilerin harici işleme çerçevelerine aktarılması ihtiyacını ortadan kaldırır. Analizi önemli ölçüde hızlandırabilir ve hesaplama yükünü azaltabilir.

3. Flexibility in Data Model: Vektör veritabanları veri modeli ve şema açısından esneklik sunar. Vektörler olarak temsil edilen metin belgeleri, resimler veya sensör okumaları gibi yapılandırılmamış veya yarı yapılandırılmış verileri işleyebilirler. Bu, karmaşık veri dönüştürmelerine veya normalleştirmeye gerek kalmadan, çeşitli veri türlerinin verimli bir şekilde depolanmasına ve alınmasına olanak tanır.

4. Scalability and Performance: Vektör veritabanları, dağıtılmış depolama ve paralel işlemeye olanak sağlayacak şekilde yatay olarak ölçeklenecek şekilde tasarlanmıştır. Bu, büyük hacimli verilerin ve yüksek sorgu yüklerinin verimli bir şekilde yönetilmesini sağlar. Vektör veritabanları, verileri birden fazla düğüme dağıtarak gerçek zamanlı uygulamalar için yüksek verim ve düşük gecikme süresi elde edebilir.

5. Machine Learning Integration: Vektör veritabanları, makine öğrenimi iş akışlarıyla sorunsuz bir şekilde bütünleşebilir. Makine öğrenimi modellerinden türetilen vektör yerleştirmelerini veya özellik temsillerini depolayabilir ve yönetebilirler. Bu, benzerlik arama, öneri veya kümeleme gibi görevler için yerleştirmelerin etkili bir şekilde alınmasına ve karşılaştırılmasına olanak tanır.

6. Real-time Analytics: Vektör veritabanları, vektör verileri üzerinde uygun ölçekte karmaşık işlemler gerçekleştirerek gerçek zamanlı analiz yetenekleri sağlayabilir. Akış verilerini işleyebilir, güncel bilgilere dayalı olarak anında analiz ve karar alma olanağı sağlarlar. Bu özellikle dolandırıcılık tespiti, anormallik tespiti veya gerçek zamanlı kişiselleştirme gibi uygulamalarda değerlidir.

7. Simplified Development and Deployment: Vektör veritabanları genellikle vektör tabanlı işlemler için uyarlanmış kullanıcı dostu arayüzler, API'ler ve sorgulama dilleri sağlar. Bu, geliştirmeyi ve mevcut uygulamalara entegrasyonu kolaylaştırır. Ek olarak, bazı vektör veritabanları bulut tabanlı yönetilen hizmetler sunarak altyapı yönetiminin operasyonel yükünü azaltır.

Ortak vektör veritabanı kullanımı:

- Görselleri görsel içeriklerine ve tarzlarına göre arayın
- Belgeleri konularına ve düşüncelerine göre arayın
- Özelliklerine ve derecelendirmelerine göre ürünleri arayın



# Hallucinations & Grounding

Büyük dil modellerinin halüsinasyon görmesi ve doğru olmayan ama doğru gibi görünen cevaplar vermesi yaygındır. Yüksek Lisans'lar internette mevcut olan neredeyse her şeyle ilgili geniş verilerle eğitildiğinden, bu durum genellikle akıl yürütmede ve uzun bağlamlı yanıtlarda kafa karışıklığına yol açar. Örneğin, halüsinasyon gören bir yüksek lisans öğrencisine Tesla'nın bu yıl elde edeceği geliri tahmin etmesi istendiğinde bunun 13,6 dolar olduğunu söyledi ki bu gerçekte olduğundan çok daha az. Yapay zeka halüsinasyonları, insan halüsinasyonlarına benzer, ancak yapay zekanın yanlış bilgileri ve gerekçesiz cevapları inandırabilecek ilginç cümleler haline getirmesi açısından farklıdır; öte yandan insan halüsinasyonları yanlış algılara dayanır. Araştırmacılar, yetersiz eğitim verileri nedeniyle oluşabilecek halüsinasyonları ve istatistiksel olaylar nedeniyle meydana gelen olumsuz halüsinasyonları belirlediler. LLM halüsinasyonlarının potansiyel nedenleri şunlardır:

1. Overgeneralization: LLM'ler eğitim verilerinden kalıplar öğrenir ve bazen bu kalıpları aşırı genelleştirme eğilimindedirler. Belirsiz veya eksik yönlendirmelerle karşı karşıya kaldıklarında, LLM'ler amaçlanan anlamın ötesine geçen yanıtlar üretebilir veya gerçekliğe dayanmayan bilgileri tahmin edebilir. Bu halüsinasyon çıktılarına yol açabilir.

2. Lack of Contextual Understanding: Yüksek Lisans'lar gerçek anlayıştan ve bağlam anlayışından yoksundur. Yanıt oluşturmak için verilerdeki istatistiksel kalıplara güvenirler. Bazı durumlarda, bağlamı yanlış yorumlayabilir veya ipucunun nüanslarını kavrayamayabilirler, bu da halüsinasyonlara veya yanlış bilgilere yol açabilir.

3. Biased Training Data: Yüksek Lisans için kullanılan eğitim verileri taraflı veya yanıltıcı bilgiler içeriyorsa, model bu önyargıları yeniden üretebilir ve güçlendirebilir. Verilerde mevcut olan önyargılar, oluşturulan çıktıları etkileyerek halüsinasyona veya ayrımcı tepkilere yol açabilir.

4. Rare or Out-of-Distribution Inputs: Yüksek Lisans'lar büyük miktarda metin verisi üzerinde eğitilir, ancak nadir veya eğitim verilerinin dağıtımının dışında olan girdilerle karşılaşabilirler. Bu tür girdilerle karşı karşıya kaldıklarında, LLM'ler uygun yanıtlar üretmekte zorlanabilir ve bunun yerine halüsinasyonlar veya anlamsız çıktılar üretebilir.

5. Unsupervised Pre-training: LLM'ler genellikle denetimsiz bir şekilde önceden eğitilir ve burada belirli bir görev denetimi olmadan çok miktarda metin verisinden öğrenirler. Bu yaklaşım onların zengin dil temsillerini öğrenmelerine olanak tanırken, aynı zamanda modelin öğrendiği kalıplara dayalı olarak hayali veya halüsinasyonlu içerik üretmesiyle de sonuçlanabilir.

6. Fine-tuning Bias: LLM'ler genellikle belirli veri kümeleri veya görevlere göre ince ayar yapılır ve bu da önyargılara veya çarpık perspektiflere neden olabilir. İnce ayar verileri önyargılıysa, halüsinasyonların veya bu önyargılarla uyumlu çıktıların oluşmasına katkıda bulunabilir.

Reinforcement learning and human Geri bildirimin, LLM'lerin temellendirilmesinde en yararlı yöntemler olduğu kanıtlanmıştır. Bu devam eden bir araştırma alanı olduğundan modellerdeki halüsinasyonları azaltabilecek hiçbir şey ortaya çıkmadı. Ustaların başka bir Ustanın doğrulaması yoluyla halüsinasyon görmesini önlemek için korkuluklar oluşturulabilir. Diğer teknikler şunları içerir:

1. Curate Diverse and Representative Training Data: Yüksek Lisans için kullanılan eğitim verilerinin çeşitli, temsili ve önyargılardan arınmış olduğundan emin olun. Veri kümesini geniş bir bakış açısı yelpazesini içerecek şekilde dikkatlice düzenleyin ve stereotipleri veya yanlış bilgileri güçlendirmekten kaçının. Dengeli bir eğitim sinyali sağlamak için birden fazla kaynaktan ve perspektiften gelen verileri birleştirmeyi düşünün.

2. Address Biases in Data: Eğitim verilerinde mevcut önyargıları tespit etmek ve azaltmak için teknikler uygulayın. Bu, önyargı analizini, veri ön işlemeyi veya artırma yöntemlerini içerebilir. Adil ve tarafsız nesli teşvik etmek için hem açık hem de örtülü önyargıları azaltmak için aktif olarak çalışın.

3. Fine-Tuning with Adversarial Examples: İnce ayar aşamasında, LLM'nin davranışına meydan okuyan düşmanca örnekleri dahil edin ve onu daha doğru ve güvenilir yanıtlar üretmeye teşvik edin. Çelişkili eğitim, modelin halüsinasyon eğilimlerini tanımlamasına ve düzeltmesine yardımcı olabilir.

4. Contextual Understanding and Reasoning: Yüksek Lisans'ın bağlamsal anlayış ve muhakeme yeteneklerini geliştirin. Yapılandırılmış bilgi sağlamak ve muhakeme yeteneklerini geliştirmek için dış bilgi kaynaklarını veya bilgi grafiklerini dahil edin. Daha doğru ve temelli yanıtlar üretmek için modeli gerçekler, mantıksal ilişkiler ve neden-sonuç kalıpları üzerinde mantık yürütmeye teşvik edin.

5. Adversarial Testing and Evaluation: Rekabetçi testleri içeren sağlam değerlendirme metodolojileri geliştirin. Modelin performansını, halüsinasyonları veya önyargıları tetiklemek için tasarlanmış zorlu istemlere veya girdilere karşı değerlendirin. Bu, zayıf yönlerin belirlenmesine yardımcı olabilir ve daha fazla iyileştirme yapılmasına rehberlik edebilir.

6. Transparent Model Explanations: Yüksek Lisans'ın şeffaflığını ve açıklanabilirliğini geliştirin. Modelin muhakeme sürecini ve dayandığı bilgi kaynaklarını vurgulayarak, oluşturulan çıktılar için açıklamalar sağlayacak yöntemler geliştirin. Bu, kullanıcıların potansiyel halüsinasyonları veya önyargıları anlamasına ve tanımlamasına yardımcı olabilir.

7. Continuous Monitoring and Iterative Improvement: Halüsinasyonları tespit etmek ve gidermek için sürekli izleme mekanizmaları oluşturun. Modelin davranışını gerçek dünya senaryolarında düzenli olarak değerlendirin ve sorunları belirleyip düzeltmek için kullanıcı geri bildirimlerini toplayın. Bu bulgulara dayanarak modeli sürekli güncelleyin ve iyileştirin.

8. Collaborative Efforts and Ethical Guidelines: Sorumlu yapay zeka gelişimine yönelik en iyi uygulamaları, araştırma bulgularını ve yönergeleri paylaşmak için yapay zeka topluluğu içinde işbirlikçi çabalara katılın. Adaleti, şeffaflığı ve kullanıcı refahını ön planda tutan etik yönergelere ve standartlara bağlılığı teşvik edin.

LLM'lerin temellendirilmesi devam eden bir zorluktur ve bilgi temsili, doğal dil anlayışı ve bilgi erişiminden oluşan tekniklerin bir kombinasyonunu gerektirir. Gerçek dünya bilgisini, gerçekleri kontrol etme mekanizmalarını ve bağlam farkındalığını birleştirerek, LLM'lerin temelini geliştirebilir ve doğru ve bağlamsal olarak uygun yanıtlar üretme konusundaki güvenilirliklerini artırabiliriz.

# Bringing Reasoning to LLMs

Karmaşık muhakeme temelli görevlerin yüksek lisans öğrencileri için üstesinden gelinmesi zordur ve sıklıkla yanlış cevaplara veya halüsinasyona yol açar. Burası SocraticAI'nin yeni yöntemleriyle LLM'lere muhakeme sağlama konusunda adım attığı yerdir: Çok Aracılı İşbirliğine Dayalı Problem Çözme (Döngüde İnsanlarla), LLM'lerin potansiyelini tam olarak açığa çıkarmak için aracılar arasında serbest biçimli sorgulama, Bilgi için 'İstek Vermeyi' Yeniden Düşünmek ve Muhakeme.

ChatGPT 3.5'te yerleşiktir. Bu çerçeve temel olarak rol yapma oyunu oynayan üç yüksek lisans temsilcisini içerir: Sokrates (Theaetetus ile etkileşime giren bir analist), Theaetetus (Sokrates ile etkileşime giren bir analist) ve Platon (alınan kararları denetleyen bir düzeltmen) sorunları işbirliği içinde çözmek için . Aracılara WolframAlpha'ya ve bir Python kod yorumlayıcısına erişim izni verilir.

Sokratik yapay zeka sistemlerinin temel özellikleri şunları içerebilir:

* Question Generation: Kullanıcıları bir konunun farklı yönlerini düşünmeye, analiz etmeye ve keşfetmeye teşvik eden sorular oluşturun. Bu sorular eleştirel düşünmeyi teşvik etmek ve daha derin anlayışı teşvik etmek için tasarlanmıştır.

* Dialogue Management: Kullanıcılarla etkileşimli görüşmelere katılın, yanıtlarına yanıt verin, daha fazla araştırma yapın ve geri bildirim veya ek istemler sağlayın. Fikirlerin yapılandırılmış ve yönlendirilmiş bir şekilde araştırılmasını kolaylaştıran bir diyalog akışını takip ederler.

* Domain Knowledge: Sorgulama süreci sırasında bağlamsal olarak ilgili bilgileri sağlamak için alana özgü bilgileri birleştirin veya harici bilgi kaynaklarına erişin. Bu, kullanıcıların tartışılan konuyu daha derinlemesine anlamalarına yardımcı olur.

* Adaptive Learning: Soru sorma stratejilerini kullanıcı yanıtlarına, başarılı sorgulamanın öğrenme kalıplarına göre uyarlamak ve deneyimi bireysel kullanıcıların ihtiyaçlarına göre uyarlamak için makine öğrenimi tekniklerini kullanın.

* Feedback and Evaluation: Kullanıcı yanıtları hakkında geri bildirimde bulunun, düşüncelerine ilişkin içgörüler sağlayın, anlayıştaki boşlukları tespit edin veya alternatif bakış açıları önerin. Ayrıca kullanıcıların akıl yürütmelerini veya çözümlerini değerlendirebilir ve iyileştirme için rehberlik sağlayabilirler.

# Responsible Generative AI

Üretken yapay zekadaki ilerlemelere rağmen riskler ve zararlar mevcut. Her yeni buluşun iyi ve kötü yanları vardır ve hangisini seçeceğimiz bize bağlıdır. Yapay zekanın zararları insanlığa pek çok açıdan pahalıya mal olabilir, bu nedenle üretken yapay zekanın önyargıyı azaltmak, adaleti teşvik etmek ve insanlığın iyiliğine hizmet edecek sorumlu sistemler haline gelmek için yönetilmesi çok önemlidir. Birçok ülke yapay zekanın yüksek lisans kullanımına yönelik düzenlenmesi konusunda adımlar atmaya başladı. Yapay zeka sistemleri etik, şeffaf, açıklanabilir, güvenilir, sürdürülebilir, güvenli ve herhangi bir kişi veya kuruluş için güvenli olmalıdır. Uygulama geliştirmenin her adımında uygun politikaların ve kalite kontrollerinin izlenmesine yardımcı olabilecek yapay zeka etiğini sağlamak için açık kaynaklı araçlar mevcuttur. Bugün neredeyse her şirket yapay zekayı kapsayıcı hale geldi ve bu da hangi verilerin kullanıldığı, hangi algoritmaların dahil edildiği, hangi değerlendirme metriklerinin ve standartlarının takip edildiği ve son olarak son kullanıcı kabulünün nasıl gerçekleştiği konusunda giderek daha dikkatli olunması sorusunu gündeme getiriyor. Deepfakes en tehlikeli yönlerden biridir ve dolandırıcılık ve siber suçları artırarak daha iyi niteliklere yükselmiştir. Yanlış bilginin yayılması, gerçek bilgilere göre daha hızlı yayılır.

1. Understand the Potential Risks: Üretken yapay zeka sistemleriyle ilgili potansiyel riskleri ve zorlukları öğrenerek başlayın. Bunlar önyargılı veya zararlı çıktıları, veri gizliliği endişelerini veya istenmeyen sonuçları içerebilir.

2. Establish Ethical Guidelines: Toplumsal değerler ve ilkelerle uyumlu bir dizi etik kural tanımlayın. Bu yönergeler, üretken yapay zeka sisteminizin geliştirilmesi ve konuşlandırılması konusunda bilgi sağlamalıdır. Bu kılavuzun geliştirilmesine etik uzmanları, alan uzmanları ve etkilenen topluluklar da dahil olmak üzere çeşitli paydaşları dahil etmeyi düşünün.

3. Data Collection and Management:  Veri toplama ve yönetim süreçlerinizin gizlilik ve izin düzenlemeleriyle uyumlu olduğundan emin olun. Eğitim verilerindeki olası önyargılara dikkat edin ve bunları azaltmak için adımlar atın. Basmakalıp yargıları güçlendirmekten veya yeterince temsil edilmeyen grupları dışlamaktan kaçınmak için çeşitli ve temsili bir veri kümesi oluşturmaya çalışın.

4. Transparency and Explainability: Üretken yapay zeka sisteminizi şeffaf ve açıklanabilir hale getirmeyi hedefleyin. Kullanıcılar sistemin nasıl çalıştığını ve hangi verileri kullandığını net bir şekilde anlamalıdır. Güven ve hesap verebilirlik oluşturmak için sistem tarafından oluşturulan kararlar veya çıktılar için açıklamalar sağlayın.

5. Bias Detection and Mitigation: Üretilen çıktılardaki önyargıları tespit etmek ve azaltmak için yöntemler uygulayın. Eğitim veya dağıtım sırasında ortaya çıkabilecek önyargıları belirlemek ve gidermek için sistemi düzenli olarak denetleyin. Kullanıcıların önyargılı çıktılar hakkında geri bildirimde bulunmasına ve sistemi buna göre iyileştirmesine olanak tanıyan mekanizmalar geliştirin.

6. User Control and Consent: Kullanıcıların, oluşturulan çıktılar üzerinde kontrol sahibi olmalarını sağlayın ve verilerinin kullanılmasına ilişkin bilgilendirilmiş onay vermelerini sağlayın. Kullanıcılara çıktıları etik sınırlar dahilinde kendi tercihlerine göre özelleştirmeleri veya değiştirmeleri için seçenekler sunun.

7. Ongoing Monitoring and Evaluation: Üretken yapay zeka sisteminizin performansını ve etkisini sürekli olarak izleyin ve değerlendirin. Etik kurallara uygunluğunu düzenli olarak değerlendirin ve ortaya çıkan sorunları veya endişeleri ele alın. Kullanıcı geri bildirimlerine duyarlı olun ve sorumlu davranışı geliştirmek için sistemi yineleyin.

8. Collaboration and Accountability: Sorumlu yapay zekanın zorluklarını toplu olarak ele almak için yapay zeka topluluğu içinde iş birliğini teşvik edin. Tartışmalara katılın, en iyi uygulamaları paylaşın ve sorumlu üretken yapay zeka sistemleri için endüstri standartları oluşturmaya çalışın. Yapay zeka teknolojilerinin sorumlu bir şekilde devreye alınmasını ve kullanılmasını sağlamak için kuruluşunuzda bir hesap verebilirlik kültürü geliştirin.
